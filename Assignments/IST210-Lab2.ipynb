##Below is my Pet Shop ER Diagram for Lab 2.

![Pet Shop ER Diagram](https://raw.githubusercontent.com/natenyahoo/IST210Stuff/master/Assignments/Lab2-PetShopERDiagram_Corrected.png)

In this ER diagram, I modeled the relationships between entities (nouns) in a pet shop on a random day. There is a lot of intricacy and moving parts that all work behind the scenes to keep things moving. For example, when a customer wants to adopt a pet, a transaction for their adoption is created, which stores the customer's loyalty program ID and the pet's ID (the pet ID must be included so that two customers can't accidentally adopt the same pet). The date of the transaction is also included, for record keeping. Interestingly, a customer can only adopt a maximum of 5 pets per transaction (as shown in the cardinality between "AdoptionTransactions" and "Pets") but a customer can make 5 adoptions a day. This means a customer could adopt a total of 5*5 = 25 guinea pigs, if they want their grass mowed by an army of guinea pigs.
